```python
#instead using "print" we will replace for RETURN to be able to save in variables:
def country_emissions(Country, key):
   
    try:
        
        n = dictionary_countries.get (Country , 'Country not found')
        print('n value: ', n)
        key = key.capitalize()
    
        if type(n) == int and key in data.keys():
            return (data[key][(272* (n-1)):(272*n)])
    
        else:
            print('The country or the key doesnt exist')
            
    except:
        print ('Something has gone wrong.')
```

```python
#The problem is we have some limitations with the year, so we will use an IF:

def global_claculation(Year):
    "'We are using this function to calculate the global emissions per year'"
    try:
        "'We filter between the min and max year'" 
        if Year >= min(set(data['Year'])) and Year <= max(set(data['Year'])):  
    
            total_global = data['Total'][Year-1750::272][-1] 
            "'In the dataset we know every 272 the country change'"
            return total_global 
            "'This is the result we need'"
        else:
            print ('The year should be between 1750 and 2021.')
    # We added some exceptions in case we have some typical issues
    except TypeError:
            print ('Please introduce a number value.')
    except ValueError:
            print ('The value cant be a text, please introduce a number value.')
    except Excepction:
            print ('Something went wrong.')
```

### Formula to check emissions per Country and per Year

```python
def country__year__source_emissions(Country, Year, Source = 'Total'):
   
    try:
        
        n = dictionary_countries.get (Country , 'Country not found')
        Year = int(Year)
        if Year >= min(data['Year']) and Year <= max(data['Year']): 
            country__year_emissions = data[Source][(272* (n-1)):(272*n)][Year-1750]
            return country__year_emissions
        else:
            print('The country or the key doesnt exist')
            
    except:
        print ('Something has gone wrong.')
```

```python
country__year__source_emissions('USA',2021, 'Cement')
```

```python
#We need a formula to know emissions in the year 1750:

for Country in list(dictionary_countries.keys())[:-1]:
    
    if country__year__source_emissions (Country, 1750) != 0.0:
        print (Country)
```
